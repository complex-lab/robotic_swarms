```python
fig,ax = plt.subplots(1,1,figsize=(12,6))
expSumFileName ='C:\\Users\\User\\Documents\\Eden\\locate_git_code\\two_circles_exp\\two_circles_exp5.csv'
expSum = pd.read_csv(expSumFileName,index_col=0)
exp_range=range(75,83)
fps=25
expSum=expSum.iloc[exp_range]
diameters_num_tot=[]
kissing_times_tot=[]
kissing_persentage_tot=[]
filename_tot=[]
framenum_tot=[]
bottype_tot=[]
sum_of_kissing_time_tot=[]
for index,row in (expSum.iterrows()):
    filename=[row.path]*3
    bottype=[row.bottype]*3
    #print(filename)
    diameters_num=[row['firstpassiveDiameter[cm]'],row['secondpassiveDiameter[cm]'],row['thirdpassiveDiameter[cm]']]
    diameters=[row['FirstDiameterLocatingFile'],row['SecondDiameterLocatingFile'],row['ThirdDiameterLocatingFile']]
#first_diameter=expSum['ThirdDiameterLocatingFile']
    kissing_times=[]
    kissing_persentage=[]
    framenums=[]
    for diameter in diameters:
        BotLocatingFileName=row.BotLocatingFile
        #print(BotLocatingFileName)
        BotLocating=pd.read_csv(BotLocatingFileName,index_col=0)
        pillar_locating_file=pd.read_csv(diameter,index_col=0)
        pillar_location=pillar_locating_file.iloc[-1]
        xbots=BotLocating.x
        ybots=BotLocating.y
        xpillar=pillar_location.x
        ypillar=pillar_location.y
        distances=(xbots-xpillar)**2+(ybots-ypillar)**2
        threshold=pillar_location.r
        rbot=np.mean(BotLocating.r)
        too_close_thresh=threshold**2
        close_threshold=(threshold+rbot+10)**2
        #print(close_threshold)
        #print(distances)
        framenum=len(distances)
        #print(np.size(distances))
        #print(framenum)
        cond=(distances<=close_threshold)
        dfC = distances[cond]
        kissing_times.append(len(dfC)/fps)
        kissing_persentage.append(len(dfC)/framenum*100)
        framenums.append(framenum)
    if row.bottype=='fronter':
        color='g'
    if row.bottype=='aligner':
        color='b'
    kissing_times=np.array(kissing_times)
    sum_of_kissing_time=np.sum(kissing_times)
    sum_of_kissing_time_array=[sum_of_kissing_time]*3
    kissing_persentage=kissing_times/sum_of_kissing_time*100
    sum_of_kissing_time_tot.extend(sum_of_kissing_time_array)
    diameters_num_tot.extend(diameters_num)
    filename_tot.extend(filename)
    kissing_times_tot.extend(kissing_times)
    kissing_persentage_tot.extend(kissing_persentage)
    framenum_tot.extend(framenums)
    bottype_tot.extend(bottype)
    ax.plot(diameters_num,kissing_persentage,'-o',color=color)
ax.set_xlabel('diamters[cm]')
ax.set_ylabel('time persentage[%]')
ax.grid()
data_tot=pd.DataFrame({'filename':filename_tot,'bottype':bottype_tot,'diameters[cm]':diameters_num_tot,'kissing_times':kissing_times_tot
                       ,'kissing_persentage':kissing_persentage_tot,'frame number':framenum_tot,'total kissing time':sum_of_kissing_time_tot})
print(data_tot)
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,6))
data_mean=data_tot.groupby(['bottype','diameters[cm]']).mean()
#print(data_mean)
data_std=data_tot.groupby(['bottype','diameters[cm]']).std()
#print(data_std)
for bottype in ['fronter','aligner']:
    current_data_mean=data_mean.loc[bottype]
    current_data_std=data_std.loc[bottype]
    if bottype=='fronter':
        color='g'
    if bottype=='aligner':
        color='b'
    ax.errorbar(current_data_mean.index,current_data_mean['kissing_persentage'],yerr=current_data_std['kissing_persentage'],color=color
                ,linestyle='-',marker='o')
ax.set_xlabel('diamters[cm]')
ax.set_ylabel('time persentage[%]')
ax.grid()
```
