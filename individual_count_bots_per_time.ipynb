```python
df=pd.read_csv('D:\\Eden\\12_1_23\\fronters-28cm-16bots\\C0570.MP4_distances.csv',index_col=0)
frame_num=pd.unique(df['frame.1'])
close_threshold=130
count=0
frame_vs_count_fronter=[]
time_fronter=[]
fps=25
for frame in frame_num:
    frame_table=df[(df['frame.1']==frame)]
    for index,row in frame_table.iterrows():
        distance=row.distance
        if distance<=close_threshold:
            count=count+1
    time_fronter.append(frame/fps)
    frame_vs_count_fronter.append(count)
    count=0
    
```
