# Package linking results for remote work

#### This sscript takes the current sate of experimetn summary csv and copy everything to a single folder
##### (this can be later added to a repository)

All data is saved in a single directory with each file having the following sturcutre
bottype_Nbots_Diam_C000.mp4
bottype_Nbots_Diam_C000_loc.csv
bottype_Nbots_Diam_C000_link.csv

```python
import pandas as pd
import os as os
```

```python
expSumFileName = 'transportExpSumMaster.csv'
expSum = pd.read_csv(expSumFileName,index_col=0)

expSum = expSum[expSum.index<42]

packageDir = 'packaged\\botsTracking'
destPackageDfFileName= packageDir + '\\expSumPack21.csv'

smallVideoSuffix = '_small.mp4'
trajBotVideoSuffix = '_BotsTraj.mp4'
```

```python
expSum
```

```python
print(expSum[expSum.index==(0-41)].locatingFileName.values)
```

```python
dfPackaged = expSum.copy()
dfPackaged['botsMiniVideoFileNamePack']=['none']*len(dfPackaged)
dfPackaged['botsLocatingFileNamePack']=['none']*len(dfPackaged)
dfPackaged['botsLinkingFileNamePack']=['none']*len(dfPackaged)

for index, row in expSum.iterrows():
    botType = row.bottype
    bots = row.bots
    diam = row["passiveDiameter[cm]"]
    head,tail = os.path.split(row.path) #get only the filename
    fileName = tail
    
    destFileNameBase = 'bots_'+botType + '_N'+str(bots).zfill(3)+'_D'+str(diam).zfill(3)+'_'+fileName
    destFileNameVideo = destFileNameBase
    destFileNameLocating = destFileNameBase+'_botsLoc.csv'
    destFileNameLinking = destFileNameBase+'_botsLink.csv'

    dfPackaged.at[index, 'botsMiniVideoFileNamePack'] = destFileNameVideo
    dfPackaged.at[index, 'botsLocatingFileNamePack']  = destFileNameLocating
    dfPackaged.at[index, 'botsLinkingFileNamePack']   = destFileNameLinking
    
    sourceFileNameLocating = row.botLocatingFileName
    sourceFileNameLinking  = row.botLinkingFileName
    sourceFileNameVideo    = row.path + smallVideoSuffix + trajBotVideoSuffix
    
    #Copy locating file
    command = 'copy ' + sourceFileNameLocating + ' ' + packageDir + '\\'+destFileNameLocating
    os.system(command)
    
    #Copy linking file
    command = 'copy ' + sourceFileNameLinking + ' ' + packageDir + '\\'+destFileNameLinking
    os.system(command)
    
    #Copy minified+traj file
    command = 'copy ' + sourceFileNameVideo + ' ' + packageDir + '\\'+destFileNameVideo
    os.system(command)
    
#save the experiment summary
dfPackaged.to_csv(destPackageDfFileName)
```
