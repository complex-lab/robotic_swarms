```python
fig,ax = plt.subplots()
for expNum in [4,38]:
    row = expSum.iloc[expNum]
    fileNameLinked = row.linkingFileName
    df = pd.read_csv(fileNameLinked)

    cond = df.particle==0
    dfC = df[cond]
    deltaFrame = dfC.frame.diff(diffConst)[diffConst:]
    deltaX = dfC.x.diff(diffConst)[diffConst:]
    deltaY = dfC.y.diff(diffConst)[diffConst:]

    deltaR = (deltaX**2+deltaY**2)**0.5
    cond = deltaFrame==diffConst
    deltaR = deltaR[cond]
    deltaFrame = deltaFrame[cond]
    ax.hist(deltaR,bins=bins,label=str(expNum),alpha=0.7);

ax.legend()
```
